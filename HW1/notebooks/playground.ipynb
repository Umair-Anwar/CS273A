```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt

# =========== 1(a) Generating Data ==================
mu = 0
sigma = 1
alpha = 0.5
dim = 2
N = 1000

mu1 = Series(np.random.normal(mu, sigma,dim))
mu2 = Series(np.random.normal(mu, sigma,dim))          
data1 = DataFrame(np.random.normal(mu,alpha*sigma,(N,dim)))+mu1
data2 = DataFrame(np.random.normal(mu,alpha*sigma,(N,dim)))+mu2


# ============= 1(b) Fisher’s linear discriminant and Evaluation ==========================
m1 = data1.mean()
m2 = data2.mean()
# Sw = data1.apply(lambda x: DataFrame(x-m1).dot(DataFrame((x-m1)).transpose(),axis=1)) + \
# 	data2.apply(lambda x: DataFrame(x-m2).dot(DataFrame((x-m2)).transpose(),axis=1))

Sw = np.zeros((dim,dim))
for i in range(len(data1)):
	Sw += DataFrame(data1.ix[i]-m1).dot(DataFrame(data1.ix[i]-m1).transpose())
	Sw += DataFrame(data2.ix[i]-m2).dot(DataFrame(data2.ix[i]-m2).transpose())

Sw_inv = inv(Sw)
w = Sw_inv.dot(m2-m1)
b = w.dot((m1+m2)/2)

label1 = data1.apply(lambda x: x.dot(w)<=b,axis=1)
label2 = data2.apply(lambda x: x.dot(w)>b,axis=1)

accuracy1 = label1.value_counts()[False]*1.0/len(label1)
accuracy2 = label2.value_counts()[False]*1.0/len(label2)

# plot in 2D
data = DataFrame(np.random.normal(mu, sigma,(N,dim)))
plt.scatter(data[0],data[1],c='gray',alpha=0.1)
plt.scatter(data1[0],data1[1],c='blue',alpha=0.5)
plt.scatter(data2[0],data2[1],c='red',alpha=0.5)
x1 = np.arange(-4,4,0.1)
x2 = -(w[0]/w[1])*x1+b/w[1]
plt.plot(x1,x2,'green')

# print result
print "accuracy1:" + str(accuracy1)
print "accuracy2:" + str(accuracy2)


```
