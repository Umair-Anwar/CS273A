```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
```

# Fisher Result Analysis

```python
Fisher = DataFrame.from_csv("./test_results/Fisher.csv")
```

```python
Fisher.reset_index('function',inplace=True)
```

### alpha & accuracy_mean

#### We can see that the accuracy_mean firstly drop when alpha grows because the overlap of two classes grow. Then it stablize in a  low value, because it can hardly discriminate between two class when alpha gets too large

```python
plt.plot(Fisher[Fisher.N==100].alpha,Fisher[Fisher.N==100].accuracy_mean,"red")
plt.plot(Fisher[Fisher.N==300].alpha,Fisher[Fisher.N==300].accuracy_mean,"yellow")
plt.plot(Fisher[Fisher.N==500].alpha,Fisher[Fisher.N==500].accuracy_mean,"blue")
plt.plot(Fisher[Fisher.N==700].alpha,Fisher[Fisher.N==700].accuracy_mean,"green")
plt.plot(Fisher[Fisher.N==900].alpha,Fisher[Fisher.N==900].accuracy_mean,"purple")
```

### alpha & accuracy_std

#### We can see that the accuracy_std first grows when alpha gets larger, because it the overlap between the two classes grow and unstable. Then we alpha gets too big, the overlap between two classes are stablized, so the accuray_std drops

```python
plt.plot(Fisher[Fisher.N==100].alpha,Fisher[Fisher.N==100].accuracy_std,"red")
plt.plot(Fisher[Fisher.N==300].alpha,Fisher[Fisher.N==300].accuracy_std,"yellow")
plt.plot(Fisher[Fisher.N==500].alpha,Fisher[Fisher.N==500].accuracy_std,"blue")
plt.plot(Fisher[Fisher.N==700].alpha,Fisher[Fisher.N==700].accuracy_std,"green")
plt.plot(Fisher[Fisher.N==900].alpha,Fisher[Fisher.N==900].accuracy_std,"purple")
```

### N & accuracy_mean

#### We can see than the accurcy_mean initially drops when N grow, because there may be more outliers. But then it become stable probabily because the outliers' number is too small compare to the total data points

```python
plt.plot(Fisher[(Fisher.alpha>0.09) & (Fisher.alpha<0.11)].N,Fisher[(Fisher.alpha>0.09) & (Fisher.alpha<0.11)].accuracy_mean,"red")
plt.plot(Fisher[(Fisher.alpha>0.29) & (Fisher.alpha<0.31)].N,Fisher[(Fisher.alpha>0.29) & (Fisher.alpha<0.31)].accuracy_mean,"yellow")
plt.plot(Fisher[(Fisher.alpha>0.49) & (Fisher.alpha<0.51)].N,Fisher[(Fisher.alpha>0.49) & (Fisher.alpha<0.51)].accuracy_mean,"blue")
plt.plot(Fisher[(Fisher.alpha>0.69) & (Fisher.alpha<0.71)].N,Fisher[(Fisher.alpha>0.69) & (Fisher.alpha<0.71)].accuracy_mean,"green")
plt.plot(Fisher[(Fisher.alpha>0.89) & (Fisher.alpha<0.91)].N,Fisher[(Fisher.alpha>0.89) & (Fisher.alpha<0.91)].accuracy_mean,"purple")
```

### N & accuracy_std

#### We can see that the accuracy_std gradually drops when N grows. This probabily because when N grows, the overlap between two classes become stable, so as the std.

```python
plt.plot(Fisher[(Fisher.alpha>0.09) & (Fisher.alpha<0.11)].N,Fisher[(Fisher.alpha>0.09) & (Fisher.alpha<0.11)].accuracy_std,"red")
plt.plot(Fisher[(Fisher.alpha>0.29) & (Fisher.alpha<0.31)].N,Fisher[(Fisher.alpha>0.29) & (Fisher.alpha<0.31)].accuracy_std,"yellow")
plt.plot(Fisher[(Fisher.alpha>0.49) & (Fisher.alpha<0.51)].N,Fisher[(Fisher.alpha>0.49) & (Fisher.alpha<0.51)].accuracy_std,"blue")
plt.plot(Fisher[(Fisher.alpha>0.69) & (Fisher.alpha<0.71)].N,Fisher[(Fisher.alpha>0.69) & (Fisher.alpha<0.71)].accuracy_std,"green")
plt.plot(Fisher[(Fisher.alpha>0.89) & (Fisher.alpha<0.91)].N,Fisher[(Fisher.alpha>0.89) & (Fisher.alpha<0.91)].accuracy_std,"purple")
```
