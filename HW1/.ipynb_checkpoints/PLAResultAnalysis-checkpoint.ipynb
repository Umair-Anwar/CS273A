```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
```

# PLA Result Analysis

```python
PLA = DataFrame.from_csv("./test_results/PLA.csv")
```

```python
PLA.reset_index('function',inplace=True)
```

### iteration & accuracy_mean

#### We can see that the accuracy_mean is stable we iteration grows. Mostly because it can converge in not much iterations

```python
plt.plot(PLA[PLA.testN==100].iteration,PLA[PLA.testN==100].accuracy_mean,"red")
plt.plot(PLA[PLA.testN==300].iteration,PLA[PLA.testN==300].accuracy_mean,"yellow")
plt.plot(PLA[PLA.testN==500].iteration,PLA[PLA.testN==500].accuracy_mean,"blue")
plt.plot(PLA[PLA.testN==700].iteration,PLA[PLA.testN==700].accuracy_mean,"green")
plt.plot(PLA[PLA.testN==900].iteration,PLA[PLA.testN==900].accuracy_mean,"purple")
```

### iteration & accuracy_std

#### We can see that the accuracy_std doesn't change much when iterations grows. Same as above, the PLA converge very soon and the classes' boundary is very clear, so the result is stable.

```python
plt.plot(PLA[PLA.testN==100].iteration,PLA[PLA.testN==100].accuracy_std,"red")
plt.plot(PLA[PLA.testN==300].iteration,PLA[PLA.testN==300].accuracy_std,"yellow")
plt.plot(PLA[PLA.testN==500].iteration,PLA[PLA.testN==500].accuracy_std,"blue")
plt.plot(PLA[PLA.testN==700].iteration,PLA[PLA.testN==700].accuracy_std,"green")
plt.plot(PLA[PLA.testN==900].iteration,PLA[PLA.testN==900].accuracy_std,"purple")
```

### testN & accuracy_mean

#### The size of test data doesn't change the accuracy too much. Because when the outliers grow the total number also grow. So the ratio doesn't change.

```python
plt.plot(PLA[PLA.iteration==100].testN,PLA[PLA.iteration==100].accuracy_mean,"red")
plt.plot(PLA[PLA.iteration==300].testN,PLA[PLA.iteration==300].accuracy_mean,"yellow")
plt.plot(PLA[PLA.iteration==500].testN,PLA[PLA.iteration==500].accuracy_mean,"blue")
plt.plot(PLA[PLA.iteration==700].testN,PLA[PLA.iteration==700].accuracy_mean,"green")
plt.plot(PLA[PLA.iteration==900].testN,PLA[PLA.iteration==900].accuracy_mean,"purple")
```

### testN & accuracy_std

#### We can see that the accuracy_std doesn't change much we testN grows. Same reason as above

```python
plt.plot(PLA[PLA.iteration==100].testN,PLA[PLA.iteration==100].accuracy_std,"red")
plt.plot(PLA[PLA.iteration==300].testN,PLA[PLA.iteration==300].accuracy_std,"yellow")
plt.plot(PLA[PLA.iteration==500].testN,PLA[PLA.iteration==500].accuracy_std,"blue")
plt.plot(PLA[PLA.iteration==700].testN,PLA[PLA.iteration==700].accuracy_std,"green")
plt.plot(PLA[PLA.iteration==900].testN,PLA[PLA.iteration==900].accuracy_std,"purple")
```
